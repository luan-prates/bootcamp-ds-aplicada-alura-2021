Primeiro eu quero importar os dados de forma dinamica, por que os arquivos estão separados por ano.

para isso tenho que listar todos os arquivos da pasta dataset.

```python
import os 
lista_arquivos = os.listdir('dataset')
```

agora preciso pegar o sufixo com o ano, que é o que varia na nomeclatura

```python
df_all = pd.DataFrame()
for arquivo in lista_arquivos:  
    df = pd.read_csv(f'dataset/{arquivo}', encoding = 'ISO-8859-1', sep=';', skiprows= 3, skipfooter=20,engine='python',error_bad_lines=False)
    ano = arquivo[28:32]
    df['Ano'] = ano
    df_all = pd.concat([df, df_all]) 
```
